## Blockchain  

**What is ?**  
Blockchain is a chain of blocks connected with cryptography.  

Each block is made of: 
* Data
* Previous Hash (Hash of the previous block - This make the connection)
* Hash (The hash of the block)

****

**HASH**  
Hash is a cryptographic code formed by:
Data, Previous hash and 'other things'.  
Bitcoin, as a sample, uses an SHA-256 algorithm which has 64 hexadecimal chars

* **SHA-256**
 * 1 - Deterministic (The same input always returns the same output)
 * 2 - Snowball (If at least a smallest part of the input changes the entire hash code suffers a big change
   * - Input 0000 0001 output FFFF FFFF FFFF FFFF 
   * - Input 0000 0002 output A8C8 893D 21HH F120A
 * 3 - Unilateral (It's not possible to get input from the output)
 * 4 - Fast (We don't want a algorithm that takes so much time to be calculated)
 * 5 - Collision support
   * Since the data can change, the previous hash can change and the 'other things' can change & since SHA is limited on 64 hex digits, it's possible (it's like 1 case out of some millions, but it's possible) that different input results in the same output, so the SHA-256 algorithm have to support this kind of collisions
   
**Immutable Registers**
In the blockchain, all registers are immutable. This is given by connection and by the Snowball in SHA-256.
Once any data of any block in the chain are changed, all the hashes would be changed too, resulting in a different chain from the block changed and forward.

**Distributed P2P**
To change a block or a chain of blocks are so difficult, but not impossible, but P2P networks can solve these problems.
Each user (computer) in the network is a node. What happens is that in a P2P network, each node has the entire structure of the blockchain (the entire chain) as a copy. So if a stranger node has a different chain this will be detected by the network and the blockchain in this strange node will be replaced by the original (by the one that is valid for the most of the nodes).

****

**Mining**
Each Block in blockchain contains a field called *Nonce* (Number that can only be used once).
The main idea here is to find with a brute-force search the correct Nonce that give us a hash that conforms to the objective (find a hash that starts with '000000' for sample).
Once this hash is found, the new block can be created and distributed to the network and the search for the next hash begins.

**Byzantine fault tolerance**
(better remember with the example of the generals)
A general sends information to all others (attack or retreat), who will respond to the information received to all other generals.
After receiving all the information, a general makes his decision based on most of the information received (attack if given more information about attacks, retreat otherwise).
The majority decision is the decision taken by the entire system.
This way, only 1 node isn't able to manipulate the decision of all the system

**Competing Chains**
As the network grows up, it's possible (and it occurs frequently) that different nodes mine a new block at the same time. This generates 2 concurrent chains. The consense protocol is to take the longest chain as the true chain, for that reason, the system will wait for the next block and then the chain who mine it first will be the new chain.
The block mined for the loser chain is called orphan block, who will be undone and all your transactions will return to the Mempool (where lies all the transactions waiting to be mined)
